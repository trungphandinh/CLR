```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
gs = pd.read_csv('/kaggle/input/genshin/genshinchar.csv')
gs.head(94)
```

```python
gs.info()
```

```python
gs.shape
```

```python
cols = list(gs.columns)
cols
```

```python
gs['gender'] = gs['model'].str.split().str[1]
gs
```

```python
list(gs['model'])
```

```python
check_null = gs.isnull().sum()
check_null
```

```python
gs = gs[gs['region'].notna()] # df[df['somecol'].notna()] -> drop rows when the value in 'somecol' is not NaN
gs = gs.reset_index(drop = True) 
gs
# reset_index: reset the indices after dropping rows with some conditions
```

```python
basic_gs = gs[cols[:13] + ["gender"]]
basic_gs
```

```python
voice_gs = gs[['character_name', 'voice_en', 'voice_jp', 'voice_cn', 'voice_kr']]
voice_gs
```

```python
ascension_gs = gs[['character_name'] + cols[17:27] + ['ascension_special_stat']]
ascension_gs
```

```python
region_count = basic_gs['region'].value_counts()
bar = region_count.plot.bar(figsize = (5, 3), color = 'skyblue', edgecolor = 'black')
for idx, val in enumerate(region_count):
    bar.text(idx, val + 0.1, str(val), ha = 'center')
plt.title('Region counting of GI characters')
plt.xlabel('Regions')
plt.ylabel('Count')
plt.show()
```

```python
pie = region_count.plot.pie(figsize = (7, 7), autopct = '%1.2f%%', startangle = 90)
plt.title('Region distribution of GI Characters')
plt.show()
```

```python
vision_count = basic_gs['vision'].value_counts()
bar = vision_count.plot.bar(figsize = (5, 3), color = 'skyblue', edgecolor = 'black')
for idx, val in enumerate(vision_count):
    bar.text(idx, val + 0.1, str(val), ha = 'center')
plt.title('Vision counting of GI characters')
plt.xlabel('Visions')
plt.ylabel('Count')
plt.show()
```

```python
pie = vision_count.plot.pie(figsize = (7, 7), autopct = '%1.2f%%', startangle = 90)
plt.title('Vision distribution of GI Characters')
plt.show()
```

```python
vis_star = basic_gs.groupby(['vision', 'star_rarity']).size().unstack(fill_value = 0)
vis_star
```

```python
vis_star.plot.bar(stacked = True)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Star rarity distribution in each vision')
plt.xlabel('Vision')
plt.ylabel('Star rarity counting')
plt.show()
```

```python
vis_gender = basic_gs.groupby(["vision", "gender"]).size().unstack(fill_value = 0)
vis_gender
```

```python
bar = vis_gender.plot.bar(stacked = True)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title("Gender in each vision")
plt.xlabel("Vision")
plt.ylabel("Gender counting")
plt.show()
```

```python
weapon = basic_gs["weapon_type"].value_counts()
bar = weapon.plot.bar(figsize = (6,4), color = 'brown', edgecolor = 'black')
for idx, val in enumerate(weapon):
    bar.text(idx, val + 0.1, str(val), ha = 'center')
plt.title("Weapon distribution")
plt.xlabel("Weapons")
plt.ylabel("Distribution")
plt.show()
```

```python
weap_vis = basic_gs.groupby(["weapon_type", "vision"]).size().unstack(fill_value = 0)
weap_vis
```
